# Larceny/Theft Are Most Common In the Afternoon & Evenings near San Francisco Hall of Justice/Country Jail/Traffic Court

The data used in this analysis looks at San Francisco Summer 2014 Incident data set.


```python
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab
import matplotlib as mpl
%matplotlib inline
import Geohash
import pandas as pd
import numpy as np
```

First we read in the the data that we will analyze and augment the data to support geospatial aggregations using a geohast and time of day categories "morning", "afternoon", and "evening".

```python
sanfranDF = pd.read_csv('sanfrancisco_incidents_summer_2014.csv')
```

Geohash is an excellent way to look at geospatial aggregrations. We calculate each geohash as well as a higher precision for (in this case 5) to support aggregation.

```python
geohashes = []

for row in sanfranDF['Location']:
    location = row.split(',')
    
    latitude = float(location[0].strip('('))
    longitude = float(location[1].strip(')'))

    geohash = Geohash.encode(latitude, longitude)    
    geohashes.append(geohash)

sanfranDF['geohash'] = geohashes
```

We then add day category to the data to do allow aggregrations by daytime. The categopries are 'morning’, ‘afternoon’, and ‘evening’ where morning is defined between the hours of 6AM and 12PM, afternoon is defined as 12PM to 7PM and evening is defined at 7PM to 6AM. In the summertime, the sun goes down ~7PM and rises ~6AM.

```python
#defining ranges for morning, afternoon and evening
l = range(0,6)
m = range(6,13)
a = range(13,20)
e = range(20,24)

timeofdays = []

for row in sanfranDF['Time']:

    iHour = int(row.split(':')[0])
    if iHour in m:
        timeofday = 'morning'
    elif iHour in a:
        timeofday = 'afternoon'
    elif iHour in e or iHour in l:
        timeofday = 'evening'
    else:
        timeofday = 'error'
    
    timeofdays.append(timeofday)

sanfranDF['TimeOfDay'] = timeofdays
sanfranDF.head()
```

```python
incidents = set(sanfranDF['Category'])

width = 2
plt.figure(figsize=(16,12))

y_pos = np.arange(len(incidents))
incident_counts = [sum(sanfranDF['Category'] == i) for i in incidents]

plt.bar(y_pos, incident_counts, align='center', alpha=0.5)
plt.xticks(y_pos, incidents, rotation=-90)
#plt.xticks(y_pos*2*width+width/2, incidents,rotation=-90)

#include the y-ticks
plt.yticks(np.arange(0,10000,500))
plt.title('San Francisco Incident Count - Summer 2014')

plt.show()
```

Wow! Bar chart shows that larceny/theft is by far the largest number of incidents occuring in the Summer of 2014 in San Franciso. So when and where are these occurring? Lets first look at time of day.

```python
larcenytheftDF = sanfranDF[sanfranDF['Category'] == "LARCENY/THEFT"]

labels = 'morning', 'afternoon', 'evening'
sizes = [sum(larcenytheftDF['TimeOfDay'] == t) for t in labels]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0, 0, 0)  # explode 1st slice

plt.pie(sizes, colors=colors, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
plt.title('Percent of Larceny/Theft Summer 2014 in San Francisco by Time Of Day')
 
plt.axis('equal')
plt.show()
```

So a high percentage of Larceny/Theft is not happening not only under the cover of night but in the afternoon as well. So is there a specific day that it is occuring on?

```python
larcenytheftDF = sanfranDF[sanfranDF['Category'] == "LARCENY/THEFT"]

labels = 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'
sizes = [sum(larcenytheftDF['DayOfWeek'] == t) for t in labels]
print sizes
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral', 'red', 'purple', 'pink']
#cmap = plt.cm.prism
#colors = cmap(np.linspace(0., 1., len(labels)))
explode = (0, 0, 0, 0)  # explode 1st slice

plt.pie(sizes, colors=colors, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
plt.title('Percent of Larceny/Theft Summer 2014 in San Francisco by Day of the Week')
 
plt.axis('equal')
plt.show()
```

Now lets look at where LARCENY/THEFT is happening. The district that the incidents are happening in is available. Let's aggregate by district and see if a district stands out.

```python
districts = set(larcenytheftDF['PdDistrict'])

plt.figure(figsize=(16,12))

y_pos = np.arange(len(districts))
district_counts = [sum(larcenytheftDF['PdDistrict'] == d) for d in districts]

plt.bar(y_pos, district_counts)
plt.xticks(y_pos, districts, rotation=-90)

plt.show()
```

The SOUTHERN district seems to have more then the others followed by CENTRAL and NORTHERN. Since we calculated the geohashes for all locations. Lets see if we can get a more refined locations where in the SOUTHERN district these are happening.

```python
larcenytheftSouthernDF = larcenytheftDF[larcenytheftDF['PdDistrict'] == "SOUTHERN"]

geohashes = list(set(larcenytheftSouthernDF['geohash']))
print "Number of geohashes in Southern district is "+str(len(geohashes))
print ""

southern_district_counts = [sum(larcenytheftSouthernDF['geohash'] == g) for g in geohashes]

southerndistrictDF = pd.DataFrame({'geohashes': geohashes, 'southerndistrict_counts': southern_district_counts})

print "The geohash with the most larceny/theft is:"
print southerndistrictDF.loc[southerndistrictDF['southerndistrict_counts'].idxmax()]
print ""
print "The top 5 geohashes for larceny/theft are:"
southerndistrictDF.sort_values('southerndistrict_counts', ascending=False).head(5)
```

Out of 650 geohashes that were cacluated as part of the SOUTHERN district, geohash '9q8yysu9cby1' by far has the most incidents of Larceny/Theft. When this is plotted, we find that this location is on Bryant Street near the Hall of Justice/Court/Jail/Police Department. A Bail Bonds is near by as well. This result seems very interesting! Could people be leaving court and then immediately committing larceny/theft on their way out?

```python
gh9q8yysu9cby1DF = larcenytheftDF[larcenytheftDF['geohash'] == "9q8yysu9cby1"]

timesofday = ['morning', 'afternoon', 'evening']

timeofday_counts = [sum(gh9q8yysu9cby1DF['TimeOfDay'] == t) for t in timesofday]

colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral', 'red', 'purple', 'pink']
explode = (0, 0, 0, 0)  # explode 1st slice

plt.pie(timeofday_counts, colors=colors, labels=timesofday, autopct='%1.1f%%', shadow=True, startangle=90)
plt.title('Percent of Larceny/Theft by time of day in San Fran Summer 2014 at geohash 9q8yysu9cby1')

plt.axis('equal')
plt.show()
```

```python
dayofweek = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

dayofweek_counts = [sum(gh9q8yysu9cby1DF['DayOfWeek'] == d) for d in dayofweek]

colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral', 'red', 'yellow', 'teal']
explode = (0, 0, 0, 0)  # explode 1st slice

plt.pie(dayofweek_counts, colors=colors, labels=dayofweek, autopct='%1.1f%%', shadow=True, startangle=90)
plt.title('Percent of Larceny/Theft by day of week in San Fran Summer 2014 at geohash 9q8yysu9cby1')

plt.axis('equal')
plt.show()
```
